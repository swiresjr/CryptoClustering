```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import calinski_harabasz_score
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
scaled_market_data = scaler.fit_transform(df_market_data[["price_change_percentage_24h","price_change_percentage_7d",
                                                   "price_change_percentage_14d","price_change_percentage_30d",
                                                   "price_change_percentage_60d","price_change_percentage_200d",
                                                   "price_change_percentage_1y"]])
```

```python
# Create a DataFrame with the scaled data
df_scaled_market_data = pd.DataFrame(scaled_market_data, columns=["price_change_percentage_24h","price_change_percentage_7d",
                                                   "price_change_percentage_14d","price_change_percentage_30d",
                                                   "price_change_percentage_60d","price_change_percentage_200d",
                                                   "price_change_percentage_1y"])

# Copy the crypto names from the original DataFrame
df_scaled_market_data["coin_id"] = df_market_data.index

# Set the coin_id column as index
df_scaled_market_data.set_index("coin_id", inplace=True)


# Display the scaled DataFrame
df_scaled_market_data
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
market_k_values = list(range(2, 12)) 
```

```python
# Create an empty list to store the inertia values
market_inertia_values = []
market_score_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in market_k_values:
    k_model = KMeans(n_clusters=i, random_state=1)
    labels = k_model.fit_predict(df_scaled_market_data)
    labels = k_model.labels_
    score = calinski_harabasz_score(df_scaled_market_data, labels)
    print(i, score)
    market_inertia_values.append(k_model.inertia_)
    market_score_values.append(score)
```

```python
# Create a dictionary for k-value scores
k_scores_market = {"k": market_k_values, "score": market_score_values}

# Create a DataFrame for the scores and display in descending order
k_scores_market_df = pd.DataFrame(k_scores_market)
k_scores_market_df.sort_values("score", ascending=False)
```

```python
# Create a dictionary with the data to plot the Elbow curve
market_elbow_curve = {"k": market_k_values, "inertia": market_inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
df_market_elbow_curve = pd.DataFrame(market_elbow_curve)
df_market_elbow_curve.sort_values("inertia", ascending=False)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_market_elbow_curve.hvplot.line(
    x="k", 
    y="inertia", 
    title="Optimal K Value for Scaled Market Data", 
    xticks=market_k_values
)
 
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for K is **4** as it occurs at the elbow point of the curve. 


---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
best_market_kvalue = 4  
model = KMeans(n_clusters=best_market_kvalue, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_scaled_market_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
predictions = model.predict(df_scaled_market_data)
df_scaled_market_data['cluster'] = predictions
# Print the resulting array of cluster values.
print(predictions)
```

```python
# Create a copy of the scaled DataFrame
df_scaled_market_data_copy = df_scaled_market_data.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_market_data_copy['cluster'] = predictions
# Display the copy of the scaled DataFrame
df_scaled_market_data_copy
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
# Create a scatter plot using hvPlot
scatter_plot_market = df_scaled_market_data_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d",   
    by="cluster",                       
    hover_cols=["coin_id"],       
    title="Cryptocurrency Changes by Market Components", 
    xlabel="24h Price Change (%)",     
    ylabel="7d Price Change (%)"       
)

# Display the plot
scatter_plot_market
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_transformed = pca_model.fit_transform(df_scaled_market_data)

# View the scaled PCA data
pca_transformed
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca_model.explained_variance_ratio_
# Print the explained variance for each principal component
print("Explained Variance for Each Principal Component:")
for i, variance in enumerate(explained_variance, start=1):
    print(f"PC{i}: {variance:.4f}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 
- PC1: 0.3487
- PC2: 0.3136
- PC3: 0.2263

```python
# Create a new DataFrame with the PCA data.
pca_transformed_df = pd.DataFrame(pca_transformed, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original DataFrame
pca_transformed_df["coin_id"] = df_market_data.index

# Set the coin_id column as index
pca_transformed_df.set_index("coin_id", inplace=True)

# Display the scaled PCA DataFrame
pca_transformed_df.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
pca_k_values = list(range(2, 12)) 
```

```python
# Create an empty list to store the inertia values
pca_inertia_values = []
pca_score_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in pca_k_values:
    k_model = KMeans(n_clusters=i, random_state=1)
    labels = k_model.fit_predict(pca_transformed_df)
    labels = k_model.labels_
    score = calinski_harabasz_score(pca_transformed_df, labels)
    print(i, score)
    pca_inertia_values.append(k_model.inertia_)
    pca_score_values.append(score)
```

```python
# Create a dictionary for k-value scores
pca_elbow_data = {"k": pca_k_values, "score": pca_score_values}

# Create a DataFrame for the scores and display in descending order
df_pca_scaled_data = pd.DataFrame(pca_elbow_data)
df_pca_scaled_data.sort_values("score", ascending=False)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_data = {"k": pca_k_values, "inertia": pca_inertia_values}

# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_data)
pca_elbow_df
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
pca_elbow_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Optimal K Value for Scaled PCA Data", 
    xticks=pca_k_values
)
```
